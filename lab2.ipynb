#### Problem 1 part B:

From reading the file candas22.txt using pd.read_csv, a ParserError shows up on line 851, which has 13 fields instead of 12 (the expected number of fields). on_bad_lines='skip' commands pandas to skip the problematic rows instead of raising an error, and continuing to read the rest of the file. Therefore, a data frame can be resulted succesfully without the rows that failed to parse as we tell pandas to ignore the offending rows.

This is not an appropriate way to load the data if we are trying to analyse the data as a whole because there will be data loss by ignoring these problematic rows and dropping them directly. This way of skipping the problematic or uncommon rows can make it harder to understand the root causes of errors. However, this could be a quick and straight-forward method to have a snapshot of the data in general without crashing into errors.

#### Problem 4 part D:

```python
def extract_address2(html_element):
    soup = BeautifulSoup(html_element, "html.parser")
    div = soup.find("div", class_="tlsmap_popup")

    if not div:
        return None
    div_text = div.string
    if not div_text:
        return None
    div_text = div_text[:-1]

    try:
        info_dict = json.loads(div_text)
    except json.JSONDecodeError:
        return None
    return info_dict("address_2", None)
```
